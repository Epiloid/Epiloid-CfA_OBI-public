```python
import pandas as pd
from functools import reduce
```

```python
from pathlib import Path
import requests

def download_csv(csv_url, filename, datadir="./data/download/"):
    """
    Download a CSV file from a URL and save it locally.

    Parameters
    ----------
    csv_url : str
        Base URL of the server
    filename : str
        Name to save the file as
    datadir : str
        Directory where the file will be saved (default: ./download/data)
    """

    datadir = Path(datadir)
    datadir.mkdir(exist_ok=True)

    filepath = datadir / filename

    response = requests.get(csv_url, headers={"Content-type": "text/csv"})

    if response.status_code == 200:
        with open(filepath, "wb") as file:
            file.write(response.content)
        print(f"CSV file downloaded and saved as {filepath}")
    else:
        print(f"Failed to download the CSV file. Status code: {response.status_code}")

    return filepath
```

## Feature Descriptions

**Ref: Comparative microelectrode array data of the functional development of hPSC-derived and rat neuronal networks**

1. < data label >_meanfiringrate_by_active_electrodes.csv: mean firing rate of the active electrodes in a well

2. < data label >_STTC.csv: spike train tiling coefficient evaluating correlations between pairs of electrodes

3. < data label >_bursts_per_min.csv: average number of bursts per minute

4. < data label >_mean_dur.csv: average burst duration in seconds for the bursts captured

5. < data label >_mean_freq_in_burst.csv: average spike frequency for the bursts captured

6. < data label >_per_spikes_in_burst.csv: percentage of spikes participating in bursts

7. < data label >_mean_spikes_in_burst.csv: average number of spikes in bursts
